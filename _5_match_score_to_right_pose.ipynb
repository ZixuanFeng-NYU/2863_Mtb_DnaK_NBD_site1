```python
import pandas as pd
df = pd.read_csv('feature_set/poses_data.csv')
```

```python
IJScore=pd.DataFrame()
from tqdm import tqdm 
Is,Js,scores=[],[],[]
for i , j in list(zip(df.ligand_id.tolist(),df.o_index.to_list())):
    lig_file='pose_site1/%d_%d.pdb'%(i,j)
    f1 = open(lig_file, 'r')
    for line in f1.readlines():
        if line.startswith('REMARK'):
            #print(line)
            score=line.split(' ')[2]
            print(score)
            print(i,j,score)
            scores.append(score)
            Is.append(i)
            Js.append(j)
return i, j, score
```

```python
IJScore['ligand_id']=Is
IJScore['o_index']=Js
IJScore['Lin_F9_Score']=scores
```

```python
Lin_F9_score=[]
for score in IJScore.Lin_F9_Score:
    s=score.split('\n')[0]
    print(s)
    Lin_F9_score.append(s)
```

```python
IJScore['Lin_F9_Score']=Lin_F9_score
```

```python
new_IJScore = IJScore.sort_values(["ligand_id", "o_index"], ascending = (True, True))
```
